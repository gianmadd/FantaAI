```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Imposta lo stile dei grafici
sns.set(style="whitegrid")
```

```python
# Percorso del file dei dati preprocessati
processed_data_path = "../data/processed/seriea/match_processed.csv"

# Carica i dati in un DataFrame
matches_df = pd.read_csv(processed_data_path)

# Visualizza le prime righe per confermare il caricamento
matches_df.head()
```

```python
# Statistiche descrittive
matches_df.describe()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=matches_df, x="matchday")
plt.title('Distribuzione delle Giornate di Gioco')
plt.xlabel('Giornata')
plt.ylabel('Numero di Partite')
plt.show()
```

```python
correlation_matrix = matches_df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title('Matrice di Correlazione')
plt.show()
```
